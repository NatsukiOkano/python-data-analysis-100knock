### ノック92：不要な文字を除外してみよう

```python
survey['comment'] = survey['comment'].str.replace('AA', '')
survey.head()

#.strip()：文字列(str)の両端の空白を削除する。
#.lstrip()：文字列(str)の左側の空白を削除する。
#.rstrip()：文字列(str)の右側の空白を削除する。
#.strip('Node')：両端の文字(Node)を削除する。
#.removeprefix('Node')：先頭の文字(Node)を削除する。
#.removesuffix('Node')：末尾の文字(Node)を削除する。
#.replace(' ', '')：文字列(str)の間の空間を削除する。
```

```python
survey['comment'] = survey['comment'].str.replace('\(.+?\)','', regex=True)
survey['comment'] = survey['comment'].str.replace('\（.+?\）','', regex=True)
survey.head()

#【基礎的なメタ文字↓】※メタ文字：正規表現(規則性のある表現)を成立させる記号。
#「\」エスケープ文字。「\」の後に続く文字の本来の意味を無効化して、ただの文字として機能させる。
#「\(, \)」グループ化の意味を持つ()を、ただの文字として機能させる。
#※エスケープ文字は、Windowsでは「\」が。Mac/Linuxでは「バックスラッシュ」として表示される。
#「.」任意の1文字。(改行を除く) 例：a.c = 'abc', 'axc'
#「*」直前の文字が0回以上繰り返す。例：a* = '', 'a', 'aaa'
#「+」直前の文字が1回以上繰り返す。例：a+ = 'a', 'aaa'
#「?」直前の文字が0回または1回。例：a? = '', 'a'
#「+?」最短一致。「+」の検出範囲を制限する。

#regex：引数。メタ文字(規則性のある記号)の設定。
#regex=True：メタ文字として機能させる。
#regex=False：メタ文字として機能させない。
```

### ノック98：顧客満足度と頻出単語の関係をみてみよう

```python
stop_words = ['時']
parts = ['名詞']
all_words = []
satisfaction = []
for n in range(len(survey)):
    text = survey['comment'].iloc[n]
    words = tagger.parse(text).splitlines()
    words_arr = []
    for i in words:
        if i == 'EOS' or i == '':
           continue
        word_tmp = i.split()[0]
        if len(i.split()) >= 4:
           part = i.split()[4].split('-')[0]
           if not (part in parts):
              continue
           words_arr.append(word_tmp)
           satisfaction.append(survey['satisfaction'].iloc[n])
    all_words.extend(words_arr)
all_words_df = pd.DataFrame({'words':all_words, 'satisfaction':satisfaction, 'count':len(all_words)*[1]})
all_words_df.head()
```

### ノック99：アンケート毎の特徴を表現してみよう

```python
#カウントベースの手法：コーパスから、自動的に効率よく、特徴量を抽出する手法。
#コーパス：自然言語処理の研究や、アプリケーション開発のために収集されたインターネット上の大量なテキスト(文字列)データ。
#特徴量：特徴を分析するために数値化したもの。

#カテゴリカル変数：カテゴリー変数。データの「区分」を表す。
#ダミー変数化：フラグ化。文字列や複数の選択肢を、[True(1)」か「False(0)」だけで表す複数の列に作り替える。
```

```python
parts = ['名詞']
all_words_df = pd.DataFrame()
satisfaction = []
for n in range(len(survey)):
    text = survey['comment'].iloc[n]
    words = tagger.parse(text).splitlines()
    words_df = pd.DataFrame()
    for i in words:
        if i == 'EOS' or i == '':
           continue
        word_tmp = i.split()[0]
        if len(i.split()) >= 4:
           part = i.split()[4].split('-')[0]
           if not (part in part):
              continue
           words_df[word_tmp] = [1]
    all_words_df = pd.concat([all_words_df, words_df], ignore_index=True)
all_words_df.head()
```

```python
all_words_df = all_words_df.fillna(0)
all_words_df.head()

#df.fillna()：「欠損値(NaN)」を()の値で埋める。
#※数値(int,float)だけでなく、文字列(str,object)や日付・時間(datetime)、真偽値(bool値)など、全てのデータ型で使用可能。
```

### ノック100：類似アンケートを探してみよう

```python
print(survey['comment'].iloc[2])
target_text = all_words_df.iloc[3]
print(target_text)
```

```python
#コサイン類似度：特徴量同士の角度の近さで類似度を表す。
#特徴量：特徴を分析するために数値化したもの。
#ダミー変数化：フラグ化。文字列や複数の選択肢を、[True(1)」か「False(0)」だけで表す複数の列に作り替える。
```

```python
import numpy as np
cos_sim = []
for i in range(len(all_words_df)):
    cos_text = all_words_df.iloc[i]
    cos = np.dot(target_text, cos_text) / (np.linalg.norm(target_text)) * np.linalg.norm(cos_text)
    cos_sim.append(cos)
all_words_df['cos_sim'] = cos_sim
all_words_df.sort_values('cos_sim', ascending=False).head()

#np.dot()：内積(スカラー積)を計算する。2つの変数に対応するベクトルの特徴量の積の和を求める。
#※ベクトル：変数とそれに対応する特徴量。
#※スカラー：数値。1つ以上の変数に対応するベクトルの特徴量の積の和。

#np.linalg.norm()：ライブラリ（numpy）に格納されるモジュール（linalg）から関数（norm）を自分のプログラムに導入する。
#ライブラリ（numpy）：「数値の塊」を操る。数学の天才。大量の数値を超高速で計算する。
#モジュール（linalg）：線形代数(ベクトル空間を研究する数学理論)のプログラムを格納したファイル。
#関数（norm）：特定のベクトルからスカラー計算。特定(1つ)の変数に対応する特徴量のスカラーを計算する。
#※ベクトル空間：直面する状況下で、ベクトルと考えられる変数とそれに対応する特徴量のこと。
#※理論：諸法則を体系化(系統的に統一)したもの。
```
