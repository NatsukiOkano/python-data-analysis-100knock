### ノック８２：映像データを読み込んでみよう

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

#モジュール（animation）：アニメーション(静止画を繋いだ動画)を作るプログラムを格納したファイル。
#クラス（FuncAnimation）：アニメーション自動再生。静止画を自動でめくる。
#ライブラリ（IPython）：Google Colabを構成するライブラリ。
#モジュール（display）：画面に画像や動画を表示(可視化)させるプログラムを格納したファイル。
#クラス（HTML）：ウェブブラウザで動画を再生するために、Python言語をHTML(ブラウザが理解できる言語)に変換する。
#ウェブブラウザ：ブラウザ。データを閲覧するためのアプリケーション。例）Google Chrome、Microsoft Edge
```

```python
# 情報取得 #
cap = cv2.VideoCapture('mov/mov01.avi')
width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
count = cap.get(cv2.CAP_PROP_FRAME_COUNT)
fps = cap.get(cv2.CAP_PROP_FPS)

print(f'画像幅: {width}')
print(f'画像高さ: {height}')
print(f'総フレーム数: {count}')
print(f'FPS: {fps}')

#cv2.VideoCapture()：動画ファイルを読み込む。
#cap.get()：動画の詳細な情報を取得する。
#cv2.CAP_PROP_FRAME_WIDTH：cap.get()の引数。画面に表示される画像の横幅。
#cv2.CAP_PROP_FRAME_HEIGHT：cap.get()の引数。画面に表示される画像の高さ。
#cv2.CAP_PROP_FRAME_COUNT：cap.get()の引数。アニメーションを構成する総フレーム数。
#cv2.CAP_PROP_FPS：cap.get()の引数。1秒間にめくる枚数を設定。アニメーションの再生速度。

#フレーム：アニメーションを構成する静止画。
```

```python
# 映像のフレーム画像化 #
num = 0
num_frame = 100
list_frame = []
while(cap.isOpened()):
    # 処理（フレームごとに切り出し）
    ret, frame = cap.read()
    # 出力（フレーム画像を書き出し）
    if ret:
       frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
       list_frame.append(frame_rgb)
       if cv2.waitKey(1) & 0xFF == ord('q'):
          break
       if num > num_frame:
          break
    num = num + 1
print('処理を完了しました')
cap.release()

#while()：()している間、ずっと繰り返す。
#cap.isOpened()：動画ファイルを読み込めたか確認する。True(はい), False(いいえ)
#while(cap.isOpened())：動画ファイルを読み込めていた(True)ら、下の処理をずっと繰り返す。
#break：実行中の繰り返し処理を強制的に終了させる。

#cap.read()：動画を再生して、1フレーム(静止画)ずつ順番に切り出す。

#cv2.cvtColor()：色の並び順や色の種類を変更する。(引数↓)
#cv2.COLOR_BGR2RGB：デフォルトの色の並び順「B(青)G(緑)R(赤)」を「R(赤)G(緑)B(青)」に並び替える。
#cv2.COLOR_BGR2GRAY：色を「白黒」に変更。
#色の並び順を変更する理由：AIやグラフの誤作動を起こさないため。
#※OpenCVは、古い世界標準(BGR)の並び順を適用しているのに対し、他のライブラリは、新しい世界標準(RGB)の並び順を適用している。
#白黒に変更する理由：AIが物体を検知しやすくするため。※物体検知の妨げとなる設定を変更することで、精度向上と高速化を図る。

#append()：指定したリストごと最後に追加する。
#extend()：指定したリストのデータのみ最後に追加する。
#【例：[1, 2, 3]に、[4, 5]を追加。↓】
# append() [1, 2, 3, [4, 5]]、extend() [1, 2, 3, 4, 5]

#if：もし～なら
#else：そうでなければ
#elif：「else + if」あるいは～なら

#cv2.waitKey(1)：1ミリ秒だけ待機。
#cv2.waitKey(0)：キーが押されるまで待機。
#0xFF = 255
#ord()：文字をASCIIコード(文字や記号を数値化したコード)に変換する。
#0xFF == ord()：「0～255」をASCIIコード(文字や記号を数値化したコード)と一致させる。
#cv2.waitKey(1) & 0xFF == ord('q')：1ミリ秒だけ待機している間に、「q」キーが押されたか確認する。

#cv2.VideoCapture()の変数.release()：読み込んだファイル(資源)の所有を解除して、他の端末(デバイス)からアクセスできる状態にする。
```

```python
# フレーム画像をアニメーションに変換 #
plt.figure()
patch = plt.imshow(list_frame[0])
plt.axis('off')
def animate(i):
    patch.set_data(list_frame[i])
anim = FuncAnimation(plt.gcf(), animate, frames=len(list_frame), interval=1000/30.0)
plt.close()

#plt.figure()：グラフを描くための「白紙」を作成する。
#figsize=(横の長さ,縦の長さ)：plt.figure()の引数。グラフの大きさを指定する。
   #※単位：インチ(1インチ = 約2.54cm)
#plt.imshow()：静止画を分析して、保存。
#plt.axis()：目盛り(x軸,y軸両方)の表示の有無を設定。
#plt.axis('on')：目盛り(x軸,y軸両方)を表示する。
#plt.axis('off')：目盛り(x軸,y軸両方)を非表示にする。

#def：「define(定義する)」の略。手順を関数として定義する。
#def trans_cost(df_tr, df_tc)：def 手順に名付けた関数名(変数1, 変数2)
#※ 関数 = 計算式。変数1, 変数2 = 定義する関数の素材。

#FuncAnimation(plt.gcf(),関数名,frames=,interval=)：アニメーションを自動再生する。
#plt.gcf()：グラフを描く白紙を指定する。
#frames：FuncAnimation()の引数。アニメーションを構成する総フレーム数。
#interval：FuncAnimation()の引数。次のフレームをめくるまで、何ミリ秒待機するかを設定。アニメーションの再生速度。

#plt.close()：グラフを描くために作成した白紙全てを記憶装置から削除する。
```

```python
# アニメーションを表示 #
HTML(anim.to_jshtml())

#HTML()：ウェブブラウザで動画を再生するために、Python言語をHTML(ブラウザが理解できる言語)に変換する。
#変数.to_jshtml()：Python言語をJavaScript(ブラウザを動かす言語)に変換する。
#HTML(anim.to_jshtml())：Python言語で作成したアニメーションをブラウザで再生できる動画プレーヤーとしてノートブック内に表示する。
```

### ノック８４：画像内のどこに人がいるのかを検出してみよう

```python
#HOG：画像から形を検出するためのアルゴリズム(実行手順)。
#HOG特徴量：HOGによって検出された形の特徴を分析するために数値化したもの。
#特徴量：特徴を分析するために数値化したもの。
```

```python
# 準備 #
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
hogParams = {'winStride':(8,8), 'padding':(32, 32), 'scale':1.05, 'hitThreshold':0, 'groupThreshold':5}

#cv2.HOGDescriptor()：画像から、HOG特徴量を検出する。
#変数.setSVMDetector()：HOGで検出した形を認識するための学習データを読み込む。
#cv2.HOGDescriptor_getDefaultPeopleDetector()：標準的な人間の形を判定するための学習データ。

#'winStride':(横方向の移動距離,縦方向の移動距離)：detectMultiScale()の引数。検出のスピードを設定。
#'winStride':(8,8)：横に8pixel,縦に8pixelずつ検出する。
#pixel(ピクセル)：画素。デジタル画像を構成する最小単位の点。RGB(赤・緑・青)の色情報を持つ。
#※1pixel：RGB(赤・緑・青)の色情報を持った1つの点。
#'padding':(横方向の余白,縦方向の余白)：detectMultiScale()の引数。画像の端の余白を設定する。
#'padding':(32, 32)：横に32pixel,縦に32pixelの余白を設定する。
#'scale':縮小率：detectMultiScale()の引数。画像の縮小率を設定する。
#'scale':1.05：画像を1.05倍(5%)ずつ縮小する。
#'hitThreshold':合格ライン：detectMultiScale()の引数。判定の合格ラインを設定する。
#'hitThreshold':0：標準的な合格ラインに設定。値を大きくするほど、誤検出が減少する。
#'groupThreshold':最低重なり個数：detectMultiScale()の引数。枠の最低重なり個数(検出の確信度)を設定。
#'groupThreshold':5：同じ場所に枠が5個以上重なっていない限り、検出しない。
```

```python
# 検出 #
img = cv2.imread('img/img01.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
human, r = hog.detectMultiScale(gray, **hogParams)
if (len(human)>0):
    for (x, y, w, h) in human:
         cv2.rectangle(img, (x,y), (x + w, y + h), (255,255,255), 3)
cv2_imshow(img)
cv2.imwrite('temp.jpg', img)

#cv2.imread()：画像ファイルを読み込む。

#cv2.cvtColor()：色の並び順や色の種類を変更する。(引数↓)
#cv2.COLOR_BGR2RGB：デフォルトの色の並び順「B(青)G(緑)R(赤)」を「R(赤)G(緑)B(青)」に並び替える。
#cv2.COLOR_BGR2GRAY：色を「白黒」に変更。
#色の並び順を変更する理由：AIやグラフの誤作動を起こさないため。
#※OpenCVは、古い世界標準(BGR)の並び順を適用しているのに対し、他のライブラリは、新しい世界標準(RGB)の並び順を適用している。
#白黒に変更する理由：AIが物体を検知しやすくするため。※物体検知の妨げとなる設定を変更することで、精度向上と高速化を図る。

#detectMultiScale()：画像の中から、特定の形を縮小度が最小になるまで繰り返し検出する。
#**(ダブルアスタリスク)変数：変数に設定した複数の引数を対象の関数に全て代入する。

#cv2.rectangle(画像,左上の座標,右下の座標,(青,緑,赤),線の太さ)：検出した形を四角で囲む。
#※左上の座標：(x,y) = x軸,y軸
#※右下の座標：(x + w, y + h) = x軸 + w(横幅), y軸 + h(高さ)
#※色情報(青,緑,赤)：最低値(0)色なし, 最高値(255)最大発色【例↓】
#(0,0,0)黒, (255,255,255)白, (0,255,255)黄色, (128,128,128)灰色
#※線の太さ：最小値(1), 最大値(上限なし), 四角の中を塗りつぶす(-1)

#cv2_imshow(変数)：読み込んだ画像ファイルをGoogle Colabのノートブック内に直接画像を表示する。
#cv2.imwrite(ファイル名,画像データ)：作成した画像に名前を付けて保存する。
```

### ノック８５：画像内の人の顔を検出してみよう

```python
# 準備 #
cascade_file = 'haarcascade_frontalface_alt.xml'
cascade = cv2.CascadeClassifier(cascade_file)

#cv2.CascadeClassifier('学習データ.xml')：xml形式の学習データを読み込む。
#'xml'：マークアップ言語(文法を定義する言語)を使って、AIの学習データを文字列として保存したファイルの拡張子(文字列)。
```

```python
# 検出 #
img = cv2.imread('img/img02.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
face_list = cascade.detectMultiScale(gray, minSize=(50, 50))

#cv2.imread()：画像ファイルを読み込む。

#cv2.cvtColor()：色の並び順や色の種類を変更する。(引数↓)
#cv2.COLOR_BGR2RGB：デフォルトの色の並び順「B(青)G(緑)R(赤)」を「R(赤)G(緑)B(青)」に並び替える。
#cv2.COLOR_BGR2GRAY：色を「白黒」に変更。
#色の並び順を変更する理由：AIやグラフの誤作動を起こさないため。
#※OpenCVは、古い世界標準(BGR)の並び順を適用しているのに対し、他のライブラリは、新しい世界標準(RGB)の並び順を適用している。
#白黒に変更する理由：AIが物体を検知しやすくするため。※物体検知の妨げとなる設定を変更することで、精度向上と高速化を図る。

#detectMultiScale()：画像の中から、特定の形を縮小度が最小になるまで繰り返し検出する。
#minSize=(横幅,高さ)：detectMultiScale()の引数。検出条件(横幅,高さ)の最小値を設定する。
#minSize=(50, 50)：横幅50pixel, 高さ50pixel未満以下は検出対象として認識しない。
```

```python
# 検出した顔に印を付ける #
for (x, y, w, h) in face_list:
     color = (0, 0, 255)
     pen_w = 3
     cv2.rectangle(img, (x,y), (x+w, y+h), color, thickness = pen_w)

#cv2.rectangle(画像,左上の座標,右下の座標,(青,緑,赤),線の太さ)：検出した形を四角で囲む。
#※左上の座標：(x,y) = x軸,y軸
#※右下の座標：(x + w, y + h) = x軸 + w(横幅), y軸 + h(高さ)
#※色情報(青,緑,赤)：最低値(0)色なし, 最高値(255)最大発色【例↓】
#(0,0,0)黒, (255,255,255)白, (0,255,255)黄色, (128,128,128)灰色
#※線の太さ：最小値(1), 最大値(上限なし), 四角の中を塗りつぶす(-1)
#thickness=：cv2.rectangle()の引数。線の太さを設定。
```

```python
cv2_imshow(img)
cv2.imwrite('temp.jpg', img)

#cv2_imshow(変数)：読み込んだ画像ファイルをGoogle Colabのノートブック内に直接画像を表示する。
#cv2.imwrite(ファイル名,画像データ)：作成した画像に名前を付けて保存する。
```

### ノック８６：画像内の人がどこに顔を向けているのかを検出してみよう

```python
import dlib
import math

#ライブラリ（dlib）：静止画の処理に特化。機械学習、顔のパーツ検出。顔の68個のパーツ(目、鼻、口など)の特定。
#ライブラリ（math）：数学の公式(実行手順)を関数に格納したライブラリ。
```

```python
# 準備 #
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
detector = dlib.get_frontal_face_detector()

#dlib.shape_predictor()：静止画の処理に特化。形状予測。過去の学習データから獲得した認識能力を汎用化させて、検出した四角い枠の形状を予測する。
#dlib.get_frontal_face_detector()：静止画の処理に特化。物体検知。四角い枠を検出する。
```

```python
# 検出 #
img = cv2.imread('img/img02.jpg')
dets = detector(img, 1)

#cv2.imread()：画像ファイルを読み込む。
#dlib.get_frontal_face_detector()の変数(デジタル画像, 拡大の段階数)
#※拡大倍率：最小値「0」標準サイズ(拡大なし), 推奨値「1」縦横2倍拡大, 最大値「上限なし」

for k, d in enumerate(dets):
    shape = predictor(img, d)

#enumerate(デジタル画像,インデックス)：インデックス(0から始まる管理番号)をデジタル画像に授ける。
#※「k」enumerate()によって授けられたindex,「d」indexが指すデータ
#dlib.shape_predictor()の変数(デジタル画像, indexが指すデータ)

    # 顔領域の表示 #
    color_f = (0, 0, 225)
    color_l_out = (225, 0, 0)
    color_l_in = (0, 255, 0)
    line_w = 3
    circle_r = 3
    fontType = cv2.FONT_HERSHEY_SIMPLEX
    fontSize = 1
    cv2.rectangle(img, (d.left(), d.top()), (d.right(), d.bottom()), color_f, line_w)
    cv2.putText(img, str(k), (d.left(), d.top()), fontType, fontSize, color_f, line_w)

#文字の書体の引数：cv2.FONT_HERSHEY_SIMPLEX(ゴシック体), cv2.FONT_HERSHEY_PLAIN(ゴシック体細字), cv2.FONT_HERSHEY_DUPLEX(ゴシック体二重線), cv2.FONT_HERSHEY_COMPLEX(明朝体)
#※アルファベット・数字専用。日本語(漢字/ひらがな/カタカナ)は対象外。

#cv2.rectangle()：静止画や動画に長方形を表示する。
#cv2.putText()：静止画や動画に文字列を表示する。

#dlib.left()：顔を囲む四角形の左のx座標。
#dlib.right()：顔を囲む四角形の右のx座標。
#dlib.top()：顔を囲む四角形の上のy座標。
#dlib.bottom()：顔を囲む四角形の下のy座標。

    # 重心を導出する箱を用意 #
    num_of_points_out = 17
    num_of_points_in = shape.num_parts - num_of_points_out
    gx_out = 0
    gy_out = 0
    gx_in = 0
    gy_in = 0
    for shape_point_count in range(shape.num_parts):
        shape_point = shape.part(shape_point_count)
        print(f'顔器官No.{shape_point_count} 座標位置: {shape_point.x},{shape_point.y}')
        # 器官ごとに描画 #
        if shape_point_count < num_of_points_out:
           cv2.circle(img,(shape_point.x, shape_point.y), circle_r, color_l_out, line_w)
           gx_out = gx_out + shape_point.x/num_of_points_out
           gy_out = gy_out + shape_point.y/num_of_points_out
        else:
           cv2.circle(img,(shape_point.x, shape_point.y), circle_r, color_l_in, line_w)
           gx_in = gx_in + shape_point.x/num_of_points_in
           gy_in = gy_in + shape_point.y/num_of_points_in

#dlib.part(番号)：顔器官が格納された68個の点(番号)を使って、特定の顔器官を抽出する。
#cv2.circle()：静止画や動画に円を描く。

    # 重心位置を描画 #
    cv2.circle(img, (int(gx_out), int(gy_out)), circle_r,(0,0,225), line_w)
    cv2.circle(img, (int(gx_in), int(gy_in)), circle_r,(0,0,0), line_w)

    # 顔の方向を計算 #
    theta = math.asin(2*(gx_in-gx_out)/(d.right()-d.left()))
    radian = theta * 180 / math.pi
    print(f'顔方向:{theta} 角度:{radian}度')

#math.sin(角度)：正弦。角度から、高さと斜辺の比率を求める。sin = 高さ÷斜辺
#math.cos(角度)：余弦。角度から、底辺と斜辺の比率を求める。cos = 底辺÷斜辺
#math.tan(角度)：正接。角度から、高さと底辺の比率を求める。tan = 高さ÷底辺
#math.asin(高さ/斜辺)：逆正弦。高さと斜辺の比率から、角度を求める。
#math.acos(底辺/斜辺)：逆余弦。底辺と斜辺の比率から、角度を求める。
#math.atan(高さ/底辺)：逆正接。高さと底辺の比率から、角度を求める。

#math.pi：円周率。少数第15位まで出力。(3.141592653589793)

    if radian < 0:
       textPrefix = 'left'
    else:
       textPrefix = 'right'
    textShow = f'{textPrefix}{round(abs(radian),1)}deg'
    cv2.putText(img, textShow, (d.left(), d.top()), fontType, fontSize, color_f, line_w)

#round()：数値を四捨五入する。
#abs()：絶対値(0から離れている数)を求める。

#cv2.putText()：静止画や動画に文字列を表示する。
```

### ノック８７：検出した情報を統合し、タイムラプスを作ってみよう

```python
#タイムラプス：一定の間隔をあけて撮影した静止画をつなげて作成された動画。
#デモ映像(デモ動画)：作成したプログラムが実際に動いている様子を第三者に確認してもらうための動画。
```

```python
print('タイムラプス生成を開始します。')

# 映像取得 #
cap = cv2.VideoCapture('mov/mov01.avi')
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#cv2.VideoCapture()：動画ファイルを読み込む。
#cap.get()：動画の詳細な情報を取得する。
#cv2.CAP_PROP_FRAME_WIDTH：cap.get()の引数。画面に表示される画像の横幅。
#cv2.CAP_PROP_FRAME_HEIGHT：cap.get()の引数。画面に表示される画像の高さ。
#cv2.CAP_PROP_FRAME_COUNT：cap.get()の引数。アニメーションを構成する総フレーム数。
#cv2.CAP_PROP_FPS：cap.get()の引数。1秒間にめくる枚数を設定。アニメーションの再生速度。
```

```python
# hog宣言 #
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
hogParams = {'winStride':(8, 8), 'padding':(32, 32), 'scale':1.05, 'hitThreshold':0, 'groupThreshold':5}

#cv2.HOGDescriptor()：画像から、HOG特徴量を検出する。
#変数.setSVMDetector()：HOGで検出した形を認識するための学習データを読み込む。
#cv2.HOGDescriptor_getDefaultPeopleDetector()：標準的な人間の形を判定するための学習データ。

#'winStride':(横方向の移動距離,縦方向の移動距離)：detectMultiScale()の引数。検出のスピードを設定。
#'winStride':(8,8)：横に8pixel,縦に8pixelずつ検出する。
#pixel(ピクセル)：画素。デジタル画像を構成する最小単位の点。RGB(赤・緑・青)の色情報を持つ。
#※1pixel：RGB(赤・緑・青)の色情報を持った1つの点。
#'padding':(横方向の余白,縦方向の余白)：detectMultiScale()の引数。画像の端の余白を設定する。
#'padding':(32, 32)：横に32pixel,縦に32pixelの余白を設定する。
#'scale'：detectMultiScale()の引数。画像の縮小率を設定する。
#'scale':1.05：画像を1.05倍(5%)ずつ縮小する。
#'hitThreshold'：detectMultiScale()の引数。判定の合格ラインを設定する。
#'hitThreshold':0：標準的な合格ラインに設定。値を大きくするほど、誤検出が減少する。
#'groupThreshold'：detectMultiScale()の引数。枠の最低重なり個数(検出の確信度)を設定。
#'groupThreshold':5：同じ場所に枠が5個以上重なっていない限り、検出しない。
```

```python
# タイムラプス作成 #
movie_name = 'timelapse.avi'
fourcc = cv2.VideoWriter_fourcc(*'XVID')
video = cv2.VideoWriter(movie_name, fourcc, 30, (width, height))

#cv2.VideoWriter_fourcc(*'圧縮形式')：動画を保存するための圧縮(縮小)形式を指定する。
# *'文字列'：文字列を1字1字の独立した文字に変換する。※OpenCVの関数fourcc()では「1文字ずつ順番に処理する」ことが規則化されているため。
#cv2.VideoWriter(ファイル名,fourcc()の変数,1秒間にめくる総フレーム数,(横幅,高さ))：動画を保存するためのファイルを作成する。

#【動画の圧縮形式↓】
# mp4v：他の端末との互換性(適合性)が高い。Windows/Mac/スマホで再生可能。
# XVID：古い環境用。圧縮効率が悪く、容量が大きい。
# avc1：高画質。圧縮効率も優れていて、容量が小さい。最新環境では主流。
# VP8：Googleが開発。ウェブブラウザとの整合性が高い。
# MJPG：動画をアニメーションとして保存。静止画の解析や編集に最適。高画質だが、圧縮効率が悪く、容量が非常に大きい。
```

```python
num = 0
while(cap.isOpened()):
      ret, frame = cap.read()
      if ret:
         if (num%10==0):
             gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
             human, r = hog.detectMultiScale(gray, **hogParams)
             if (len(human)>0):
                 for (x, y, w, h) in human:
                      cv2.rectangle(frame, (x, y), (x + w, y + h), (255,255,255), 3)
             video.write(frame)
         else:
             break
         num = num + 1

#while()：()している間、ずっと繰り返す。
#cap.isOpened()：動画ファイルを読み込めたか確認する。True(はい), False(いいえ)
#while(cap.isOpened())：動画ファイルを読み込めていた(True)ら、下の処理をずっと繰り返す。
#break：実行中の繰り返し処理を強制的に終了させる。

#cap.read()：動画を再生して、1フレーム(静止画)ずつ順番に切り出す。

#cv2.cvtColor()：色の並び順や色の種類を変更する。(引数↓)
#cv2.COLOR_BGR2RGB：デフォルトの色の並び順「B(青)G(緑)R(赤)」を「R(赤)G(緑)B(青)」に並び替える。
#cv2.COLOR_BGR2GRAY：色を「白黒」に変更。
#色の並び順を変更する理由：AIやグラフの誤作動を起こさないため。
#※OpenCVは、古い世界標準(BGR)の並び順を適用しているのに対し、他のライブラリは、新しい世界標準(RGB)の並び順を適用している。
#白黒に変更する理由：AIが物体を検知しやすくするため。※物体検知の妨げとなる設定を変更することで、精度向上と高速化を図る。

#detectMultiScale()：画像の中から、特定の形を縮小度が最小になるまで繰り返し検出する。
#**(ダブルアスタリスク)変数：変数に設定した複数の引数を対象の関数に全て代入する。

#cv2.rectangle(画像,左上の座標,右下の座標,(青,緑,赤),線の太さ)：検出した形を四角で囲む。
#※左上の座標：(x,y) = x軸,y軸
#※右下の座標：(x + w, y + h) = x軸 + w(横幅), y軸 + h(高さ)
#※色情報(青,緑,赤)：最低値(0)色なし, 最高値(255)最大発色【例↓】
#(0,0,0)黒, (255,255,255)白, (0,255,255)黄色, (128,128,128)灰色
#※線の太さ：最小値(1), 最大値(上限なし), 四角の中を塗りつぶす(-1)

#cv2.VideoWriter_fourcc(*'圧縮形式')：動画を保存するための圧縮(縮小)形式を指定する。
# *'文字列'：文字列を1字1字の独立した文字に変換する。※OpenCVの関数fourcc()では「1文字ずつ順番に処理する」ことが規則化されているため。
#cv2.VideoWriter(ファイル名,fourcc()の変数,1秒間にめくる総フレーム数,(横幅,高さ))：動画を保存するためのファイルを作成する。
#cv2.VideoWriterの変数.write()：作成したファイルに動画を保存する。

#【動画の圧縮形式↓】
# mp4v：他の端末との互換性(適合性)が高い。Windows/Mac/スマホで再生可能。
# XVID：古い環境用。圧縮効率が悪く、容量が大きい。
# avc1：高画質。圧縮効率も優れていて、容量が小さい。最新環境では主流。
# VP8：Googleが開発。ウェブブラウザとの整合性が高い。
# MJPG：動画をアニメーションとして保存。静止画の解析や編集に最適。高画質だが、圧縮効率が悪く、容量が非常に大きい。
```

### ノック９０：移動平均を計算することでノイズの影響を除去しよう

```python
import numpy as np
def moving_average(x, y):
    y_conv = np.convolve(y, np.ones(5)/float(5), mode='valid')
    x_dat = np.linspace(np.min(x), np.max(x), np.size(y_conv))
    return x_dat, y_conv

#numpy ：「数値の塊」を操る。数学の天才。大量の数値を超高速で計算する。
#pandas：「データフレーム(表)の構造」を操る。例）並べ替え、抽出、結合、欠損値の穴埋め、CSVの読み書きなど

#def：「define(定義する)」の略。手順を関数として定義する。
#def trans_cost(df_tr, df_tc)：def 手順に名付けた関数名(変数1, 変数2)
#※ 関数 = 計算式。変数1, 変数2 = 定義する関数の素材。

#np.convolve()：畳み込み積分で移動平均を求める。= index(行番号)が近い要素(データ)をグループ化して、その中の平均値を持った要素をグループの代表値とする。
#np.ones()：「1」を要素に持つSeries(一次元リスト)を作成する。
#np.ones(5) = [1, 1, 1, 1, 1]
#np.linspace(開始値,終了値,要素値)：開始値から終了値までを要素数で等間隔に分割した、数値のNumPy配列(ndarray)を作成する。
#np.linspace(0,100,5) = [0,25,50,75,100], np.linspace(0,100,2) = [0, 100]
#np.min()：Python標準のリスト, PandasのSeries/DataFrame, NumPy配列(ndarray)から最小値を抽出する。
#np.max()：Python標準のリスト, PandasのSeries/DataFrame, NumPy配列(ndarray)から最大値を抽出する。
#np.size()：Python標準のリスト, PandasのSeries/DataFrame, NumPy配列(ndarray)に格納される要素数を数える。

#return：結果を返して終了する。
```
