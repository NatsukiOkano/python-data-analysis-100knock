### ノック71：人間関係のネットワークを可視化してみよう

```python
import pandas as pd
df_links = pd.read_csv('links.csv', index_col='Node')
df_links.head()

#index_col：どの変数を「行ラベル兼index」として使うかを指定する。
#※実行すると、元のindex(行番号)は削除される。

#Node：点。頂点。
#None：値が存在しない。
```

```python
import networkx as nx
import matplotlib.pyplot as plt

#ライブラリ（networkx）：ネットワーク構造を計算・操作・分析する。
#ライブラリ（matplotlib）：全ての工程を自らの手で制御する「プロ仕様版」グラフ作成ツール。
#ライブラリ（matplotlib.pyplot）：複雑な工程を自動化した「簡易操作版」グラフ作成ツール。
```

```python
G = nx.Graph()

# 頂点(点)の設定
NUM = len(df_links.index)
for i in range(NUM):
    node_no = df_links.columns[i].strip('Node')
    G.add_node(str(node_no))

# 辺の設定
for i in range(NUM):
    for j in range(NUM):
        node_name = 'Node' + str(j)
        if df_links[node_name].iloc[i]==1:
           G.add_edge(str(i), str(j))

nx.draw_networkx(G, node_color='k', edge_color='k', font_color='w')
plt.show()

#nx.Graph()：ネットワーク図のデータ(オブジェクト)を作成する。

#.strip()：文字列(str)の両端の空白を削除する。
#.lstrip()：文字列(str)の左側の空白を削除する。
#.rstrip()：文字列(str)の右側の空白を削除する。
#.strip('Node')：両端の文字(Node)を削除する。
#.removeprefix('Node')：先頭の文字(Node)を削除する。
#.removesuffix('Node')：末尾の文字(Node)を削除する。
#.replace(' ', '')：文字列(str)の間の空間を削除する。

#add_node()：作成したネットワーク図に「新しい頂点(点)」を追加する。
#add_edge()：作成したネットワーク図に「新しい辺(edge)」を追加する。

#nx.draw()：設定したデータ(オブジェクト)をネットワーク図に反映させる。
#node_color：頂点(点)の色を設定。※ K = black(黒)
#edge_color：辺(線)の色を設定。
#font_color：頂点(点)ラベルの色を設定。※ w = white(白)

#plt.show()：作成したネットワーク図を表示(可視化)する。
```

### ノック72：口コミによる情報伝播の様子を可視化してみよう

```python
#伝播：活性化すること。広く伝わること。拡散。
```

```python
import numpy as np

#ライブラリ（numpy）：「数値の塊」を操る。数学の天才。大量の数値を超高速で計算する。
```

```python
def determine_link(percent):
    rand_val = np.random.rand()
    if rand_val <= percent:
      return 1
    else:
      return 0

#def：「define(定義する)」の略。手順を関数として定義する。
#def trans_cost(df_tr, df_tc)：def 手順に名付けた関数名(変数1, 変数2)
#※ 関数 = 計算式。変数1, 変数2 = 定義する関数の素材。

#np.random.rand()：「0以上1未満」の乱数を設定。
#np.random.seed(1)：乱数の発生パターン(型)を「1」が名前の設定に固定。
#乱数：規則性がなく、予測できない数字のこと。

#if：もし～なら
#else：そうでなければ
#elif：「else + if」あるいは～なら

#return：結果を返して終了する。
#return 1：出力失敗。異常あり(1)。エラー。
#return 0：出力成功。異常なし(0)。結果を返して終了する。
```

```python
def simulate_percolation(num, list_active, percent_percolation):
    for i in range(num):
        if list_active[i]==1:
           for j in range(num):
               node_name = 'Node' + str(j)
               if df_links[node_name].iloc[i]==1:
                  if determine_link(percent_percolation)==1:
                     list_active[j] = 1
    return list_active

#num：変数。※ num = number(数値)
#return 変数：関数の中から結果を反映させた変数を取り出す。
#※関数の要素となった変数は、「return df」で取り出さなければ表示されない。
#※関数の中で起きたことは、関数の外からは見えない。
```

```python
percent_percolation = 0.1
T_NUM = 36
NUM = len(df_links.index)
list_active = np.zeros(NUM)
list_active[0] = 1

list_timeSeries = []
for t in range(T_NUM):
    list_active = simulate_percolation(NUM, list_active, percent_percolation)
    list_timeSeries.append(list_active.copy())

#np.zeros()：要素(データ)が全て「0」の表またはリストを新しく作成する。

#【pandasの基本的なデータ構造(2つ)↓】
#Series：一次元リスト(列)。全ての要素にindex(行番号)が付いている。例）'A':[1, 2, 3], 'B':[4, 5, 6]
#DataFrame：二次元リスト。Seriesの集合体。

#append()：指定したリストごと最後に追加する。
#extend()：指定したリストのデータのみ最後に追加する。
#【例：[1, 2, 3]に、[4, 5]を追加。↓】
# append() [1, 2, 3, [4, 5]]、extend() [1, 2, 3, 4, 5]

#df.copy()：複製。コピーして新しいデータフレーム(表)を作る。
```

```python
def active_node_coloring(list_active):
    print(list_timeSeries[t])
    list_color = []
    for i in range(len(list_timeSeries[t])):
        if list_timeSeries[t][i]==1:
           list_color.append('r')
        else:
           list_color.append('k')
    print(len(list_color))
    return list_color
```

### ノック74：会員数の時系列変化をシミュレーションしてみよう

```python
def simulate_population(num, list_active, percent_percolation, percent_disapparence, df_links):
    # 拡散(口コミの伝播) #
    for i in range(num):
        if list_active[i]==1:
           for j in range(num):
               node_name = 'Node' + str(j)
               if df_links[node_name].iloc[i]==1:
                  if determine_link(percent_percolation)==1:
                     list_active[j] = 1

    # 消滅(利用の中断) #
    for i in range(num):
        if determine_link(percent_disapparence)==1:
           list_active[i] = 0
    return list_active

#for i in range：「繰り返し処理」の基本セット
  # for：～の間、繰り返す。
  # i：今、何回目か数えるための変数。
  # in range()：繰り返す回数。
```

```python
percent_percolation = 0.1
percent_disapparence = 0.05
T_NUM = 100
NUM = len(df_links.index)
list_active = np.zeros(NUM)
list_active[0] = 1

#np.zeros()：要素(データ)が全て「0」の表またはリストを新しく作成する。
```

```python
list_timeSeries = []
for t in range(T_NUM):
    list_active = simulate_population(NUM, list_active, percent_percolation, percent_disapparence, df_links)
    list_timeSeries.append(list_active.copy())
```

### ノック75：パラメータの全体像を相図を見ながら把握しよう

```python
#パラメータ：変数間を仲立ちする変数。
#相図：状態図。温度や圧力などの条件によって、変わる物質の状態(個体・液体・気体)をグラフ化したもの。
```

```python
print('相図計画開始')
T_NUM = 100
NUM_PhaseDiagram = 20
phaseDiagram = np.zeros((NUM_PhaseDiagram, NUM_PhaseDiagram))
for i_p in range(NUM_PhaseDiagram):
    for i_d in range(NUM_PhaseDiagram):
        percent_percolation = 0.05 * i_p
        percent_disapparence = 0.05 * i_d
        list_active = np.zeros(NUM)
        list_active[0] = 1
        for t in range(T_NUM):
            list_active = simulate_population(NUM, list_active, percent_percolation, percent_disapparence, df_links)
            phaseDiagram[i_p][i_d] = sum(list_active)
print(phaseDiagram)
```

```python
plt.matshow(phaseDiagram)
plt.colorbar(shrink=0.8)
plt.xlabel('pervent_dispparence')
plt.ylabel('percent_percolation')
plt.xticks(np.arange(0.0, 20.0, 5), np.arange(0.0, 1.0, 0.25))
plt.yticks(np.arange(0.0, 20.0, 5), np.arange(0.0, 1.0, 0.25))
plt.tick_params(bottom=False, left=False, right=False, top=False)
plt.show()

#plt.matshow()：二次元グラフのヒートマップ
#plt.colorbar()：ヒートマップの凡例
#plt.colorbar(shrink=0.8)：ヒートマップの凡例の大きさを「0.8倍」にする。
#plt.xlabel()：x軸(横軸)ラベル
#plt.ylabel()：y軸(縦軸)ラベル
#plt.xticks()：x軸(横軸)の目盛り
#plt.yticks()：y軸(縦軸)の目盛り
#np.arange(開始,終了,間隔)：開始から終了までの数値を一定の間隔で自動生成する。
#plt.tick_params(bottom(下), left(左), right(右), top(上))：目盛りを表示させる位置
#※True(表示), False(非表示)
```

### ノック78：シミュレーションのために実データからパラメータを推定しよう

```python
NUM = len(df_mem_info.index)
T_NUM = len(df_mem_info.columns)-1
```

```python
# 消滅(利用の中断)の確率推定 #
count_active = 0
count_active_to_inactive = 0
for t in range(T_NUM):
    for i in range(NUM):
        if (df_mem_info.iloc[i,t]==1):
            count_active_to_inactive += 1
            if (df_mem_info.iloc[i,t+1]==0):
                count_active += 1
estimated_percent_disapparence = count_active/count_active_to_inactive

# 拡散(口コミの伝播)の確率推定 #
count_link = 0
count_link_to_active = 0
count_link_temp = 0
for t in range(T_NUM):
    df_link_t = df_mem_info[df_mem_info[str(t)]==1]
    temp_flag_count = np.zeros(NUM)
    for i in range(len(df_link_t.index)):
        index_i = int(df_link_t.index[i])
        df_link_temp = df_mem_links[df_mem_links[f'Node{index_i}']==1]
        for j in range(len(df_link_temp.index)):
            index_j = int(df_link_temp.index[j])
            if (df_mem_info.iloc[index_j,t]==0):
                if (temp_flag_count[index_j]==0):
                    count_link += 1
                if (df_mem_info.iloc[index_j,t+1]==1):
                    if (temp_flag_count[index_j]==0):
                        temp_flag_count[index_j] = 1
                        count_link_to_active += 1
estimated_percent_percolation = count_link_to_active/count_link
```
